## FILE HANDLING

```python
f = open('abc.txt','w')
print('File name : ',f.name)
print('File Mode : ',f.mode)
print(f.readable())
print(f.writable())
print(f.closed)
f.close()
print(f.closed)
```

```python
with open("abc.txt","r") as file:
    content = file.read()
print(content)     
```

```python
with open("abc.txt","r") as file:
    content = file.readline()
print(content)     
```
